```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
var = pd.read_csv('../../data/2020/20_lap_var.csv')
yellow = pd.read_csv('../../data/2020/20_driver_yellow.csv')
```

```python
df = var.merge(yellow, on=['Driver'], how='inner')
```

```python
z = list(zip(df['Var'], df['Yellow']))
z = sorted(z, key=lambda t: t[0])
    
X = [t[0] for t in z]
X = np.array(X).reshape(-1, 1)
y = [t[1] for t in z]

clf = LogisticRegression(random_state=0).fit(X, y)
yg = clf.predict_proba(X)[:, 1]

plt.title('Lap Variance versus Prob. of Yellow Flag')
plt.plot(X, yg)
plt.show()
```
